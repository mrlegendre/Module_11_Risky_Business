```python
import numpy as np
import pandas as pd
from pathlib import Path
from collections import Counter
```

```python
from sklearn.metrics import balanced_accuracy_score
from sklearn.metrics import confusion_matrix
from imblearn.metrics import classification_report_imbalanced
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
# Load the data
file_path = Path('/Resources/LoanStats_2019Q1.csv')
df = pd.read_csv(file_path)

# Preview the data
df.head()
```

```python
# Create our features
X = df.drop(columns="loan_status")
X=pd.get_dummies(X)

# Create our target
y = df["loan_status"]
```

```python
X.describe()
```

```python
 # Check the balance of our target values
y.value_counts()
```

```python
# Split the X and y into X_train, X_test, y_train, y_test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, 
                                                   y, 
                                                   random_state=1, 
                                                   stratify=y)
X_train.shape
```

```python
# Create the StandardScaler instance
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
# Fit the Standard Scaler with the training data
# When fitting scaling functions, only train on the training dataset
X_scaler = scaler.fit(X_train)
```

```python
# Scale the training and testing data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
 # Resample the training data with the BalancedRandomForestClassifier
from imblearn.ensemble import BalancedRandomForestClassifier
classifier = BalancedRandomForestClassifier(n_estimators=100, random_state=1)
classifier.fit(X_test_scaled, y_test)
```

```python
 # Calculated the balanced accuracy score
from sklearn.metrics import balanced_accuracy_score
y_balanced = classifier.predict(X_test_scaled)
balanced_accuracy_score(y_test, y_balanced)
```

```python
# Display the confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_balanced)
```

```python
 # Print the imbalanced classification report
from imblearn.metrics import classification_report_imbalanced

print(classification_report_imbalanced(y_test, y_balanced))
```

```python
 # List the features sorted in descending order by feature importance
importances = classifier.feature_importances_
importances_sorted = sorted(zip(classifier.feature_importances_, X.columns), reverse=True)
importances_sorted
```

```python
# Train the Classifier
from imblearn.ensemble import EasyEnsembleClassifier
classifier_easy = EasyEnsembleClassifier(n_estimators=100, random_state=1)
classifier_easy.fit(X_test_scaled, y_test)
```

```python
 # Calculated the balanced accuracy score
y_easy = classifier_easy.predict(X_test_scaled)
balanced_accuracy_score(y_test, y_easy)
```

```python
# Display the confusion matrix
y_easy_pred = classifier_easy.predict(X_test_scaled)
confusion_matrix(y_test, y_easy_pred)
```

```python
 # Print the imbalanced classification report
print(classification_report_imbalanced(y_test, y_easy_pred))
```
